```python
import matplotlib as plot
import seaborn as sns
!pip install plotly
import plotly
print ('Install complete')
#imported data visualization packages
```

```python
cut_bins = [0,1,2]
bin_cut = pd.cut(df['Rnd'], cut_bins).value_counts(sort=False)
#testing out matplotlib
```

```python
bin_cut
#testing out matplotlib
```

```python
dist = bin_cut.to_frame().reset_index()
#testing out matplotlib
```

```python
dist
#testing out matplotlib
```

```python
dist['Round']=dist['index'].astype(str)
#testing out matplotlib
```

```python
import plotly.express as px
#testing out matplotlib
```

```python
fig=px.bar(dist,x="Round",y="Rnd",text="Rnd")
#testing out matplotlib
```

```python
fig.show()
#testing out matplotlib
```

```python
sns.set_theme(style='darkgrid')

df = pd.read_csv(r'C:\Users\evaldez\Desktop\Data Science Projects\NFL Draft 2018 PB.csv')
df = df.groupby(['Rnd']).sum().reset_index()

df = sns.catplot(
    data=df, kind="bar",
    x='Rnd', y='PB')

df.set_axis_labels('Round', 'Pro Bowlers')
#testing out seaborn
```
